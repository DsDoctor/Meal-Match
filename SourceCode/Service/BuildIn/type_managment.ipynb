```python
import pandas as pd
import os

BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath('')))
```

```python
data = pd.read_excel(os.path.join(BASE_DIR, 'DAO/data/ingredients.xlsx'))
```

```python
data
```

```python
ing_types = data.groupby('type').size()
ing_types
```

```python
for t in ing_types:
    print(t)
```

```python
with open(os.path.join(BASE_DIR, 'DAO/data/ing_type.txt'), 'w') as f:
    for t in ing_types:
        f.write(t + '\n')
```
